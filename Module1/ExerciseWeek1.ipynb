```python
# Bài 1: Viết function thực hiện đánh giá classification model bằng F1-
def exercise1(tp, fp, fn):
    try:
        if type(tp) != int:
            raise TypeError("TP must be int")
        elif type(fp) != int:
            raise TypeError("FP must be int")
        elif type(fn) != int:
            raise TypeError("FN must be int")
        if tp <= 0 or fp <= 0 or fn <= 0:
            raise ValueError("TP and FP and FN must be greater than zero")
        precision = tp / (tp + fp)
        recall = tp / (tp + fn)
        f1_score = 2 * ((precision * recall) / (precision + recall))
        print("precision is %.2f" %precision)
        print("recall is %.2f" %recall)
        print("f1_score is %.2f" %f1_score)
        return None
    except Exception as e:
        print(str(e))
        return None
        
# exercise1(2, 3, 4)
# exercise1('a', 3, 4)
# exercise1(2, 'a', 4)
# exercise1(2, 3, 0)
# exercise1(2.1, 3, 0)
# # print("Precision nè: %.2f" %precision)
# # print("Recall nè: %.2f" %recall)
# # print("F1_Score nè: %.2f" %f1_score)
```

```python
# Bài 2: Viết function mô phỏng theo 3 activation function.
import math
def is_number(x):
    try:
        if x.isalpha() :
            raise TypeError("Vui lòng nhập một số thực!!!")
        return True
    except Exception as e:
        print(str(e))
        return False
    
def sigmoid(x):
    x = float(x)
    return 1 / (1 + pow(math.e, -x))
def relu(x):
    x = float(x)
    if x <= 0:
        return 0
    return x
def elu(x, a):
    x = float(x)
    if x <= 0:
        return a * (pow(math.e, x) - 1)
    return x

def exercise2():
    x = input("Input x = ")
    if is_number(x) == False:
        return None
    elif is_number(x) == True:
        key = input("Nhập key đi bé: ")
        if key == 'sigmoid':
            print(sigmoid(x))
        elif key == 'relu':
            print(relu(x))
        elif key == 'elu':
            print(elu(x, float(input("Nhập a: "))))
        else:
            print("%s is not supported!!!" %key)

exercise2()
```

```python
# Viết function lựa chọn regression loss function để tính loss:
import random 
import sys
def mae(n, target, predict):
    return 1/n * abs(target - predict)

def mse(n, target, predict):
    return 1/n * pow(target - predict, 2)

def rmse(n, target, predict):
    return pow(mse(n, target, predict), 1/2)

def choice(key, n, target, predict):
    if key == 'MAE':
        return mae(n, target, predict)
    elif key == 'MSE':
        return mse(n, target, predict)
    elif key == 'RMSE':
        return rmse(n, target, predict)

num_samples = input()
if num_samples.isnumeric() == False:
    print('number of samples must be an integer number')
    exit
elif num_samples.isnumeric() == True:
    num_samples = int(num_samples)
    key = input("Nhập loss name đi em: ")
    # dic = {'MAE': 'MAE', 'RMSE': 'RMSE', 'MSE': 'MSE'}
    # if key != dic.keys:
    #     print('Lost name is not found')
    #     exit()
    for i in range(0, num_samples+1, 1):
        print(i)
        target = random.random() * 10
        predict = random.random() * 10
        if i == num_samples:        
            print("Final %s: %f" %(key, float(choice(key, num_samples, target, predict))))
        elif i != num_samples:
            print("loss name: %s, sample: %d, pred: %f, target: %f, loss: %f" %(key, num_samples, predict, target, float(choice(key, num_samples, target, predict))))

```

```python
# 4. Viết 4 functions để ước lượng các hàm số sau.
import math

def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

def approx_sin(x, n):
    sin_x = 0
    for i in range(n):
        term = ((-1) ** i * x ** (2 * i + 1)) / factorial(2 * i + 1)
        sin_x += term
    return sin_x

def approx_cos(x, n):
    cos_x = 0
    for i in range(n):
        term = ((-1) ** i * x ** (2 * i)) / factorial(2 * i)
        cos_x += term
    return cos_x

def approx_sinh(x, n):
    sinh_x = 0
    for i in range(n):
        term = (x ** (2 * i + 1)) / factorial(2 * i + 1)
        sinh_x += term
    return sinh_x

def approx_cosh(x, n):
    cosh_x = 0
    for i in range(n):
        term = (x ** (2 * i)) / factorial(2 * i)
        cosh_x += term
    return cosh_x
```

```python
# 5. Viết function thực hiện Mean Difference of n
# th Root Error:
def md_nre_single_sample(y, y_hat, n, p):
    root_y = y ** (1 / n)
    root_y_hat = y_hat ** (1 / n)
    return abs(root_y - root_y_hat) ** p

def md_nre(y_list, y_hat_list, n, p):
    m = len(y_list)
    total = 0
    for y, y_hat in zip(y_list, y_hat_list):
        total += md_nre_single_sample(y, y_hat, n, p)
    return total / m
```
